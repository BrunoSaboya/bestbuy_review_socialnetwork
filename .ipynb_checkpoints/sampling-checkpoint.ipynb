```python
import os
import pandas as pd
import json
from collections import Counter
import seaborn as sns
from netpixi.integration.gt import *
from netpixi.integration.gt import Graph
from regression.integration.gt import *
import netpixi
from collections import defaultdict
from itertools import combinations
# g = Graph(directed=False) # não-dirigido
# Get a list of all files in the directory
files = os.listdir('./jsons')

# Filter the list to only JSON files
json_files = [file for file in files if file.endswith('.json')]
```

```python
# Initialize an empty DataFrame
df_combined = pd.DataFrame()

# Loop over the JSON files and concatenate them into the DataFrame
for json_file in json_files:
    data = []
    with open(f"jsons/{json_file}", 'r', encoding='utf-8') as f:
        for line in f:
            try:
                data.append(json.loads(line))
            except json.JSONDecodeError:
                pass
    df = pd.DataFrame(data)
    df_combined = pd.concat([df_combined, df], ignore_index=True)
print(len(df_combined))
```

```python
usuarios = []
for reviews_produto in df_combined['reviews']:
    for review in reviews_produto:
        if 'username' in review.keys():
            usuarios.append(review['username'])
```

```python
data = []
for i, row in df_combined.iterrows():
    for review in row['reviews']:
        if 'username' in review.keys() and 'rating' in review.keys():
            username = review['username']
            rating = review['rating']
            product_name = row['name']
            brand = row['brand']
            data.append({'product_name': product_name, 'brand': brand, 'username': username, 'rating': rating})

new_df = pd.DataFrame(data)
print(new_df.head())
```

```python
print(len(new_df))
num_duplicates = new_df.duplicated(keep=False).sum()
print(f"Number of duplicated rows: {num_duplicates}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
from collections import defaultdict
import itertools

# Create a dictionary of user to products reviewed
user_to_products = defaultdict(list)

for i, row in new_df.iterrows():
    product_id = row['product_name']
    username = row['username']
    user_to_products[username].append(product_id)
```

```python
# Create a dictionary where the key is a pair of products and the value is the count of common reviewers
product_pairs = defaultdict(int)

for products in user_to_products.values():
    for product1, product2 in itertools.combinations(sorted(products), 2):
        product_pairs[(product1, product2)] += 1

# Extract the counts (weights) and plot a histogram
weights = list(product_pairs.values())

plt.figure(figsize=(10,6))
sns.histplot(weights, bins=30, kde=False)
plt.title('Distribution of shared reviewers between pairs of products')
plt.xlabel('Number of shared reviewers (weight)')
plt.ylabel('Number of product pairs')
plt.show()
```

```python
# Create a dictionary where the keys are usernames and the values are lists of products they've reviewed
user_products = defaultdict(list)

# Create a dictionary to map product names to unique numeric identifiers
product_ids = {}
next_id = 0

for i, row in new_df.iterrows():
    product_name = row['product_name']
    username = row['username']
    
    # If this is the first time we're seeing this product, assign it a new ID
    if product_name not in product_ids:
        product_ids[product_name] = next_id
        next_id += 1
    
    # Add the product's ID to the user's list of reviewed products
    user_products[username].append(product_ids[product_name])
```

```python
# Generate the edge list
edges = set() # using set to avoid duplicates
for products in user_products.values():
    # For each pair of products reviewed by the same user, add an edge
    for product1, product2 in combinations(sorted(products), 2):
        edges.add((product1, product2))
```

```python
# Write the edge list to a file
with open('edges.txt', 'w') as f:
    for edge in sorted(edges):  # sorted to maintain order
        f.write(f'{edge[0]}\t{edge[1]}\n')
```
